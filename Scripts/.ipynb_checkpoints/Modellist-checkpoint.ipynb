```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 读取数据
df = pd.read_csv('totaldownload_1807.csv')

# 数据预处理
df_downloads = df[df['downloads_all_time'] != 0]
df_likes = df[df['likes'] != 0]

log_data_downloads = np.log(df_downloads['downloads_all_time'])
log_data_likes = np.log(df_likes['likes'])

# 画图
plt.figure(figsize=(20, 10))

# 绘制下载次数的分布图
plt.hist(log_data_downloads, bins=50, edgecolor='k', alpha=0.6, label='Downloads')

# 绘制点赞次数的分布图
plt.hist(log_data_likes, bins=50, edgecolor='k', alpha=0.6, label='Likes', color='orange')

# 添加标题和标签，并设置字体大小
plt.title('Distribution of log(Downloads) and log(Likes)', fontsize=30)
plt.xlabel('Log Value', fontsize=26)
plt.ylabel('Frequency', fontsize=26)
plt.tick_params(axis='both', which='major', labelsize=26)

# 添加图例
plt.legend(fontsize=20)

# 显示图像
plt.show()
```

```python
import pandas as pd

# 加载Excel文件
file_path = 'sorted_downloads.xlsx'
xls = pd.ExcelFile(file_path)

# 读取每个Sheet的数据
df_last_30_days = pd.read_excel(xls, 'Last 30 Days')
df_all_time = pd.read_excel(xls, 'All Time')
df_likes = pd.read_excel(xls, 'Likes')

# 设置'name'列为索引以便于更新
df_last_30_days.set_index('name', inplace=True)
df_all_time.set_index('name', inplace=True)
df_likes.set_index('name', inplace=True)

# 使用 All Time 表中的数据更新 Last 30 Days 和 Likes 表
df_last_30_days.update(df_all_time)
df_likes.update(df_all_time)

# 重置索引
df_last_30_days.reset_index(inplace=True)
df_likes.reset_index(inplace=True)

# 将更新后的数据保存回新的Excel文件
with pd.ExcelWriter('updated_sheets.xlsx') as writer:
    df_last_30_days.to_excel(writer, sheet_name='Last 30 Days', index=False)
    df_all_time.reset_index().to_excel(writer, sheet_name='All Time', index=False)
    df_likes.to_excel(writer, sheet_name='Likes', index=False)

print("处理完成，结果已保存到文件 'updated_sheets.xlsx'")
```

```python
import pandas as pd
from collections import Counter

# Load the Excel file
file_path = 'updated_sheets_new.xlsx'
excel_data = pd.ExcelFile(file_path)

# Load each sheet into a DataFrame
df_last_30_days = pd.read_excel(file_path, sheet_name='Last 30 Days')
df_all_time = pd.read_excel(file_path, sheet_name='All Time')
df_likes = pd.read_excel(file_path, sheet_name='Likes')

# Fill missing values and standardize string format
fill_values = {
    'datasets': '未提供',
    'eval_results': '未提供',
    'license': '未指定',
    'Github': '无'
}

df_last_30_days.fillna(fill_values, inplace=True)
df_all_time.fillna(fill_values, inplace=True)
df_likes.fillna(fill_values, inplace=True)

# Function to check missing values in eval_results column
def check_missing_eval_results(df):
    missing_count = df['eval_results'].isnull().sum()
    total_count = df.shape[0]
    missing_percentage = (missing_count / total_count) * 100
    return missing_count, missing_percentage

missing_last_30_days = check_missing_eval_results(df_last_30_days)
missing_all_time = check_missing_eval_results(df_all_time)
missing_likes = check_missing_eval_results(df_likes)

print("Missing eval_results (Last 30 Days):", missing_last_30_days)
print("Missing eval_results (All Time):", missing_all_time)
print("Missing eval_results (Likes):", missing_likes)

# Function to find the most frequent words in eval_results column
def most_frequent_words(df, num_words=10):
    eval_texts = df['eval_results'].dropna().str.lower().apply(lambda x: set(x.split()))
    all_words = [word for text in eval_texts for word in text]
    most_common_words = Counter(all_words).most_common(num_words)
    return most_common_words

freq_words_last_30_days = most_frequent_words(df_last_30_days)
freq_words_all_time = most_frequent_words(df_all_time)
freq_words_likes = most_frequent_words(df_likes)

print("Most frequent words in eval_results (Last 30 Days):", freq_words_last_30_days)
print("Most frequent words in eval_results (All Time):", freq_words_all_time)
print("Most frequent words in eval_results (Likes):", freq_words_likes)
```

```python
import pandas as pd
from collections import Counter

# Load the Excel file
file_path = 'updated_sheets_new.xlsx'
excel_data = pd.ExcelFile(file_path)

# Load each sheet into a DataFrame
df_last_30_days = pd.read_excel(file_path, sheet_name='Last 30 Days')
df_all_time = pd.read_excel(file_path, sheet_name='All Time')
df_likes = pd.read_excel(file_path, sheet_name='Likes')

# Fill missing values
fill_values = {
    'datasets': '未提供',
    'eval_results': '未提供',
    'license': '未指定',
    'Github': '无'
}

df_last_30_days.fillna(fill_values, inplace=True)
df_all_time.fillna(fill_values, inplace=True)
df_likes.fillna(fill_values, inplace=True)

# Standardize string format
def standardize_strings(df, columns):
    for col in columns:
        df[col] = df[col].str.strip().str.lower()
    return df

string_columns = ['name', 'author', 'datasets', 'eval_results', 'license', 'Github', 'HuggingFace_repo']
df_last_30_days = standardize_strings(df_last_30_days, string_columns)
df_all_time = standardize_strings(df_all_time, string_columns)
df_likes = standardize_strings(df_likes, string_columns)

# Check missing values in eval_results column
def check_missing_eval_results(df):
    missing_count = (df['eval_results'] == '未提供').sum()
    total_count = df.shape[0]
    missing_percentage = (missing_count / total_count) * 100
    return missing_count, missing_percentage

missing_last_30_days = check_missing_eval_results(df_last_30_days)
missing_all_time = check_missing_eval_results(df_all_time)
missing_likes = check_missing_eval_results(df_likes)

print("\n### Missing eval_results ###")
print("Missing eval_results (Last 30 Days):", missing_last_30_days)
print("Missing eval_results (All Time):", missing_all_time)
print("Missing eval_results (Likes):", missing_likes)

# Find most frequent words in eval_results column
def most_frequent_words(df, num_words=10):
    eval_texts = df['eval_results'].dropna().str.lower().apply(lambda x: set(x.split()))
    all_words = [word for text in eval_texts for word in text]
    most_common_words = Counter(all_words).most_common(num_words)
    return most_common_words

freq_words_last_30_days = most_frequent_words(df_last_30_days)
freq_words_all_time = most_frequent_words(df_all_time)
freq_words_likes = most_frequent_words(df_likes)

print("\n### Most Frequent Words in eval_results ###")
print("Most frequent words in eval_results (Last 30 Days):", freq_words_last_30_days)
print("Most frequent words in eval_results (All Time):", freq_words_all_time)
print("Most frequent words in eval_results (Likes):", freq_words_likes)

# Count values in openness_train_code column
def count_openness_train_code(df):
    return df['openness_train_code'].value_counts()

openness_train_code_last_30_days = count_openness_train_code(df_last_30_days)
openness_train_code_all_time = count_openness_train_code(df_all_time)
openness_train_code_likes = count_openness_train_code(df_likes)

print("\n### Openness Train Code ###")
print("Openness Train Code (Last 30 Days):\n", openness_train_code_last_30_days)
print("Openness Train Code (All Time):\n", openness_train_code_all_time)
print("Openness Train Code (Likes):\n", openness_train_code_likes)

# Check if Github link exists
def check_github_existence(df):
    github_exists = (df['Github'] != '无').sum()
    github_not_exists = (df['Github'] == '无').sum()
    return github_exists, github_not_exists

github_last_30_days = check_github_existence(df_last_30_days)
github_all_time = check_github_existence(df_all_time)
github_likes = check_github_existence(df_likes)

print("\n### Github Existence ###")
print("Github existence (Last 30 Days):", github_last_30_days)
print("Github existence (All Time):", github_all_time)
print("Github existence (Likes):", github_likes)

# Check openness_dataset values
def check_openness_dataset(df):
    openness_0 = (df['openness_dataset'] == 0).sum()
    openness_1 = (df['openness_dataset'] == 1).sum()
    return openness_0, openness_1

openness_dataset_last_30_days = check_openness_dataset(df_last_30_days)
openness_dataset_all_time = check_openness_dataset(df_all_time)
openness_dataset_likes = check_openness_dataset(df_likes)

print("\n### Openness Dataset ###")
print("Openness Dataset (Last 30 Days):", openness_dataset_last_30_days)
print("Openness Dataset (All Time):", openness_dataset_all_time)
print("Openness Dataset (Likes):", openness_dataset_likes)

# Count the frequency of different licenses
def license_analysis(df):
    return df['license'].value_counts()

license_last_30_days = license_analysis(df_last_30_days)
license_all_time = license_analysis(df_all_time)
license_likes = license_analysis(df_likes)

print("\n### License Analysis ###")
print("License (Last 30 Days):\n", license_last_30_days)
print("License (All Time):\n", license_all_time)
print("License (Likes):\n", license_likes)

# Find records where openness_train_code is 1, openness_dataset is 1, and eval_results is not '未提供'
def find_specific_records(df):
    filtered_df = df[(df['openness_train_code'] == 1) & (df['openness_dataset'] == 1) & (df['eval_results'] != '未提供')]
    return filtered_df

specific_records_last_30_days = find_specific_records(df_last_30_days)
specific_records_all_time = find_specific_records(df_all_time)
specific_records_likes = find_specific_records(df_likes)

print("\n### Specific Records (Last 30 Days) ###")
specific_records_last_30_days
# print("\n### Specific Records (All Time) ###")
# print(specific_records_all_time)

# print("\n### Specific Records (Likes) ###")
# print(specific_records_likes)
```
